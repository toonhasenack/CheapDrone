```python
import pygame
import socket

host='localhost'
port=8091

pygame.init()
joysticks = []
clock = pygame.time.Clock()

joystick = pygame.joystick.Joystick(0)
joystick.init()
print ("Detected joystick "),joystick.get_name(),"'"

throttle = 128
turn = 128
leftRight = 128
forwardBackward = 128
command = 0

with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as sock:
    while True:
        clock.tick(60)
        for event in pygame.event.get():
            #check for joystick action
            if event.type == 1536:
                if event.dict["axis"] == 0:
                    turn = min(int(128 + 128*event.dict["value"]),255)
                elif event.dict["axis"] == 1:
                    throttle = min(int(128 - 128*event.dict["value"]),255)
                elif event.dict["axis"] == 2:
                    leftRight = min(int(128 + 128*event.dict["value"]),255)
                elif event.dict["axis"] == 3:
                    forwardBackward = min(int(128 - 128*event.dict["value"]),255)
                else:
                    pass
            
            #check for button press
            elif event.type == 1539:
                if event.button == 0:
                    command = 1
                elif event.button == 1:
                    command = 2
                elif event.button == 3:
                    command = 3

            elif event.type == 1540:
                command = 0
            
            message = bytearray()
            for part in [leftRight, forwardBackward, throttle, turn, command]:
                message.append(part)
            sock.sendto(message, (host, port))
            print(f"Sent: {message}")
```
